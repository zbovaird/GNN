```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes with types of devices
G.add_node("Internet", shape="cloud")
G.add_node("Perimeter FortiGate Firewall", shape="box")
G.add_node("Core FortiSwitch 1", shape="ellipse")
G.add_node("Core FortiSwitch 2", shape="ellipse")
G.add_node("Internal FortiGate Firewall", shape="box")
G.add_node("Access FortiSwitch 1", shape="ellipse")
G.add_node("Access FortiSwitch 2", shape="ellipse")
G.add_node("Access FortiSwitch 3", shape="ellipse")
G.add_node("Access FortiSwitch 4", shape="ellipse")
G.add_node("PLC 1", shape="triangle")
G.add_node("PLC 2", shape="triangle")
G.add_node("PLC 3", shape="triangle")
G.add_node("PLC 4", shape="triangle")
G.add_node("SCADA Server", shape="hexagon")
G.add_node("HMI", shape="hexagon")
G.add_node("FortiAnalyzer", shape="parallelogram")
G.add_node("FortiManager", shape="parallelogram")
G.add_node("FortiNAC", shape="parallelogram")
G.add_node("FortiSIEM", shape="parallelogram")
G.add_node("FortiAP 1", shape="octagon")
G.add_node("FortiAP 2", shape="octagon")
G.add_node("FortiAP 3", shape="octagon")

# Add edges to represent connections
G.add_edges_from([
    ("Internet", "Perimeter FortiGate Firewall"),
    ("Perimeter FortiGate Firewall", "Core FortiSwitch 1"),
    ("Perimeter FortiGate Firewall", "Core FortiSwitch 2"),
    ("Core FortiSwitch 1", "Internal FortiGate Firewall"),
    ("Core FortiSwitch 2", "Internal FortiGate Firewall"),
    ("Internal FortiGate Firewall", "Access FortiSwitch 1"),
    ("Internal FortiGate Firewall", "Access FortiSwitch 2"),
    ("Internal FortiGate Firewall", "Access FortiSwitch 3"),
    ("Internal FortiGate Firewall", "Access FortiSwitch 4"),
    ("Access FortiSwitch 1", "PLC 1"),
    ("Access FortiSwitch 2", "PLC 2"),
    ("Access FortiSwitch 3", "PLC 3"),
    ("Access FortiSwitch 4", "PLC 4"),
    ("Core FortiSwitch 1", "SCADA Server"),
    ("Core FortiSwitch 2", "HMI"),
    ("Core FortiSwitch 1", "FortiAnalyzer"),
    ("Core FortiSwitch 2", "FortiManager"),
    ("Core FortiSwitch 1", "FortiNAC"),
    ("Core FortiSwitch 2", "FortiSIEM"),
    ("Access FortiSwitch 1", "FortiAP 1"),
    ("Access FortiSwitch 2", "FortiAP 2"),
    ("Access FortiSwitch 3", "FortiAP 3"),
])

# Define positions for nodes
pos = {
    "Internet": (0, 3),
    "Perimeter FortiGate Firewall": (1, 3),
    "Core FortiSwitch 1": (2, 3),
    "Core FortiSwitch 2": (2, 2),
    "Internal FortiGate Firewall": (3, 2.5),
    "Access FortiSwitch 1": (4, 3.5),
    "Access FortiSwitch 2": (4, 2.5),
    "Access FortiSwitch 3": (4, 1.5),
    "Access FortiSwitch 4": (4, 0.5),
    "PLC 1": (5, 3.5),
    "PLC 2": (5, 2.5),
    "PLC 3": (5, 1.5),
    "PLC 4": (5, 0.5),
    "SCADA Server": (2.5, 4),
    "HMI": (2.5, 1),
    "FortiAnalyzer": (2.5, 3.5),
    "FortiManager": (2.5, 0.5),
    "FortiNAC": (2.5, 2.75),
    "FortiSIEM": (2.5, 1.75),
    "FortiAP 1": (4.5, 3.75),
    "FortiAP 2": (4.5, 2.75),
    "FortiAP 3": (4.5, 1.75),
}

# Draw the graph
plt.figure(figsize=(14, 10))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue', font_size=10, font_weight='bold', arrowsize=20)
plt.title("Basic Fortinet System Layout for an ICS Plant")
plt.show()
```
